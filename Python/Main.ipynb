```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/Users/dam124/Desktop/Unsupervised/Superstore.csv')
data
```

Droping Columns

```python
data = data.drop(['Country', 'Row ID'], axis=1)
```

```python
data.describe()
```

```python
product_sales = data.groupby('Product Name')['Sales'].sum()

top_products = product_sales.sort_values(ascending=False).head(5)
percent_sales = 100 * top_products / top_products.mean()
plt.pie(percent_sales, labels=percent_sales.index, autopct='%1.1f%%')
plt.title('Top Selling Products')
plt.show()
```

```python
product_sales = data.groupby('Product Name')['Sales'].sum()

bottom_products = product_sales.sort_values().head(5)

percent_sales = 100 * bottom_products / bottom_products.sum()


plt.pie(percent_sales, labels=percent_sales.index, autopct='%1.1f%%')

plt.title('5 Least Selling Products')


plt.show()
```

```python
df = pd.DataFrame(data)

df.plot(kind='scatter', x='Discount', y='Sales', color='red', s=30)

plt.xlabel('Discount')
plt.ylabel('Total Sales')
plt.title('Discount and Sales Relationship ')
plt.show()
```

```python
##Top Customers Based On Sales

customer_sales = data.groupby('Customer Name')['Sales'].sum()

top_customers = customer_sales.sort_values(ascending=False).head(10)
print(top_customers)
```

```python
subcategory_profit = data.groupby('Sub-Category')['Profit'].mean()
subcategory_profit = subcategory_profit.sort_values(ascending=False)[:5] 

subcategory_profit.plot(kind='bar', figsize=(10, 6))

plt.title('Top 5 Sub-Categories by Average Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Average Profit')

plt.show()
```

```python
subcategory_profit = data.groupby('Sub-Category')['Profit'].mean()
subcategory_profit = subcategory_profit.sort_values(ascending=True)[:5] # Select subcategories with least profit

subcategory_profit.plot(kind='bar', figsize=(10, 6))

plt.title('Sub-Categories with the Least Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Average Profit')

plt.show()
```

```python

import seaborn as sns

data['Order_Year'] = pd.to_datetime(data['Order_Year'], format='%Y')
subcategory_profit = data.groupby(['Sub-Category', data['Order_Year'].dt.year])['Profit'].mean().reset_index()
least_profitable_subcategories = subcategory_profit.groupby('Sub-Category')['Profit'].mean().nsmallest(5).index
subcategory_profit = subcategory_profit[subcategory_profit['Sub-Category'].isin(least_profitable_subcategories)]
sns.lineplot(x='Order_Year', y='Profit', hue='Sub-Category', data=subcategory_profit)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.title('Year And The Sub-Category With Least Profit')
plt.xlabel('Order Year')
plt.ylabel('Average Profit')

plt.show()
```

```python
grouped = data.groupby("Segment")

segment_counts = grouped.size().reset_index(name='counts')

plt.bar(segment_counts['Segment'], segment_counts['counts'])

plt.xlabel("Segment")
plt.ylabel("Count")
plt.title("Segment Distribution")
plt.show()
```

```python
grouped = data.groupby("Ship Mode")

segment_counts = grouped.size().reset_index(name='counts')

plt.bar(segment_counts['Ship Mode'], segment_counts['counts'])

plt.xlabel("Ship Mode")
plt.ylabel("Number Of Sales")
plt.title("Distribution Of Ship Mode")
plt.show()
```

```python
category_counts = data['Category'].value_counts()

sns.barplot(x=category_counts.index, y=category_counts)


plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution Of Categories')


plt.show()
```

```python
df = pd.DataFrame(data)
df['Difference'] = df['Ship_Day'] - df['Order_Day']
grouped_df = df.groupby('Ship Mode').mean()
plt.bar(grouped_df.index, grouped_df['Difference'])
plt.xlabel('Ship Mode')
plt.ylabel('Average Shipping Days')

plt.title('Shiping Time (Days)')

plt.show()
```

```python
grouped_df = df.groupby(['Ship Mode'])['Profit'].mean().reset_index()

sns.barplot(x='Ship Mode', y='Profit', data=grouped_df)
plt.xlabel('Ship Mode')
plt.ylabel('Average Profit')
plt.title('Average Profit Per Ship Mode')
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.DataFrame(data)
df.groupby('Ship Mode')['Sales'].mean().plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Average Sales')
plt.title('Average Sales Per Ship Mode')
plt.show()
```

```python
grouped = df.groupby('City').sum()


grouped['Average Profit'] = df.groupby('City').sum()['Profit']
grouped = grouped.nlargest(25, 'Profit')
# Plot the data
grouped.plot(kind='bar', y='Profit', color='skyblue', legend=False)

plt.xlabel('City')
plt.ylabel('Total Profit')
plt.title('Cities with High Profits')
plt.xticks(rotation=90)

plt.show()
```

```python
grouped_df = df.groupby(['Order_Month']).mean()



plt.plot(grouped_df.index, grouped_df['Sales'])

plt.title('Monthly Sales')
plt.xlabel('Month')

plt.ylabel('Average Sales')

plt.show()
```

```python
df = pd.DataFrame(data)
df.groupby('Region')['Sales'].mean().plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Average Sales')
plt.title('Average Of Sales Per Region')
plt.show()
```

```python
df = pd.DataFrame(data)
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.title('Sum Of Sales Per Region')
plt.show()
```

```python

top_5_discounts_by_region = df.groupby('Region')['Discount'].nlargest(1)

# Plot the top 5 discounts for each region using a bar chart
top_5_discounts_by_region.plot(kind='bar', figsize=(10, 6))
plt.title('Top 5 Discounts by Region')
plt.xlabel('Region')
plt.ylabel('Discount In Percentage')
plt.show()

```
