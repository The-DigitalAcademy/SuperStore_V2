```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Superstore.csv')
data
```

Droping Columns

```python
data = data.drop(['Country', 'Row ID'], axis=1)
```

```python
data.isnull()
```

```python
data.duplicated()
```

```python
data.describe()
```

```python
product_sales = data.groupby('Product Name')['Sales'].mean()

top_products = product_sales.sort_values(ascending=False).head(5)
percent_sales = 100 * top_products / top_products.mean()
plt.pie(percent_sales, labels=percent_sales.index, autopct='%1.1f%%')
plt.title('Top Selling Products')
plt.show()
```

```python
product_sales = data.groupby('Product Name')['Sales'].sum()

bottom_products = product_sales.sort_values().head(5)

percent_sales = 100 * bottom_products / bottom_products.sum()


plt.pie(percent_sales, labels=percent_sales.index, autopct='%1.1f%%')

plt.title('5 Least Products')


plt.show()
```

```python
df = pd.DataFrame(data)


df.plot(kind='scatter', x='Discount', y='Sales', color='red', s=30)


plt.xlabel('Discount')
plt.ylabel('Sales')
plt.title('Discount vs Sales ')


plt.show()
```

```python
##Top Customers Based On Sales

customer_sales = data.groupby('Customer Name')['Sales'].sum()

top_customers = customer_sales.sort_values(ascending=False).head(10)
print(top_customers)
```

```python
subcategory_profit = data.groupby('Sub-Category')['Profit'].mean()


subcategory_profit = subcategory_profit.sort_values(ascending=False)


subcategory_profit.plot(kind='bar', figsize=(10, 6))

plt.title('Sub-Category Average Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Mean Profit')

plt.show()
```

```python
grouped = data.groupby("Segment")

segment_counts = grouped.size().reset_index(name='counts')

plt.bar(segment_counts['Segment'], segment_counts['counts'])

plt.xlabel("Segment")
plt.ylabel("Count")
plt.title("Popular Segment")
plt.show()
```

```python
grouped = data.groupby("Ship Mode")

segment_counts = grouped.size().reset_index(name='counts')

plt.bar(segment_counts['Ship Mode'], segment_counts['counts'])

plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.title("Preffered Ship Mode")
plt.show()
```

```python
unique_categories = data['Category'].unique()


print(unique_categories)
```

```python

df = pd.DataFrame(data)
df['Difference'] = data['Ship_Day'] - df['Order_Day']
grouped_df = df.groupby('Ship Mode').mean()
plt.bar(grouped_df.index, grouped_df['Difference'])
plt.xlabel('Ship Mode')
plt.ylabel('Difference (Ship_Day - Order_Day)')

plt.title('Average Ship Days')

plt.show()
```

```python
grouped_df = data.groupby([ 'Sub-Category'])['Profit'].mean().reset_index()

plt.scatter(grouped_df['Sub-Category'], grouped_df['Profit'])
plt.xlabel('Sub-Category')
plt.ylabel('Average Profit')
plt.xticks(rotation=90)
plt.title('Average Profit Per Sub-Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

grouped_df = data.groupby(['Ship Mode'])['Profit'].mean().reset_index()

sns.lineplot(x='Ship Mode', y='Profit', data=grouped_df)
plt.xlabel('Ship Mode')
plt.ylabel('Mean Profit')
plt.title('Average Profit Per Ship Mode')
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.DataFrame(data)
df.groupby('Ship Mode')['Sales'].mean().plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Sales')
plt.title('Average Sales Per Ship Mode')
plt.show()
```

```python
grouped = data.groupby('City').sum()


grouped['Average Profit'] = data.groupby('City').mean()['Profit']
grouped = grouped.nlargest(25, 'Profit')
# Plot the data
grouped.plot(kind='bar', y='Profit', color='skyblue', legend=False)

plt.xlabel('City')
plt.ylabel('Total Profit')
plt.title('Cities with High Profits')
plt.xticks(rotation=90)

plt.show()
```

```python
grouped_df = data.groupby(['Order_Month']).mean()

plt.plot(grouped_df.index, grouped_df['Sales'])
plt.title('Monthly Sales')

plt.xlabel('Month')

plt.ylabel('Sales')

plt.show()
```
